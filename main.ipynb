```python
# Import thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Bước 1: Dữ liệu
X = np.array([[0, 0], [1, 2], [2, 1], [1, 1], [1, 0], [0, 1]])  # Tập dữ liệu
y = np.array([1, 1, 1, -1, -1, -1])  # Nhãn tương ứng (Class 1: 1, Class 2: -1)

# Bước 2: Khởi tạo mô hình SVM với kernel RBF
model = SVC(kernel='rbf')

# Bước 3: Huấn luyện mô hình
model.fit(X, y)

# Bước 4: In hệ số b (intercept)
print("Hệ số b (intercept):", model.intercept_)

# Bước 5: Vẽ đồ thị biểu diễn SVM
def plot_svm_decision_boundary(X, y, model):
    # Tạo lưới để vẽ các điểm
    h = .02  # Độ phân giải
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Dự đoán trên lưới
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Vẽ ranh giới quyết định
    plt.contourf(xx, yy, Z, alpha=0.8)

    # Vẽ các điểm dữ liệu
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', s=100)
    plt.xlabel('Đặc trưng 1')
    plt.ylabel('Đặc trưng 2')
    plt.title('SVM Decision Boundary (Kernel RBF)')
    plt.show()

# Vẽ đồ thị
plot_svm_decision_boundary(X, y, model)
```
